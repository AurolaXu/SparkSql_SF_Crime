```
from pyspark.sql import SparkSession
spark = SparkSession \
    .builder \
    .appName("crime analysis") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()

df_opt1 = spark.read.format("csv").option("header", "true").load(data_path)
display(df_opt1)
df_opt1.createOrReplaceTempView("sf_crime")
```

```
from pyspark.sql.functions import hour, date_format, to_date,month,year
# X to Longitude, Y to latitude in doubles and Date to date format
SF_DT = df_opt1.withColumn('Longitude', df_opt1['X'].cast('double')) \
                 .withColumn('Latitude', df_opt1['Y'].cast('double')) \
                 .withColumn("Crime Date",to_date(df_opt1.Date, "MM/dd/yyyy")) 
display(SF_DT)
```

```
# SF downtown spacial range: X (-122.4213,-122.4313), Y(37.7540,37.7740). 
# Spark dataframe based solution for Q3
q3_result = SF_DT.filter((SF_DT.DayOfWeek == "Sunday") & (SF_DT.Longitude>=-122.4313) & (SF_DT.Longitude<= -122.4213)& (SF_DT.Latitude>= 37.7540 ) & (SF_DT.Latitude<=37.7740) ).groupby('Crime Date','DayOfWeek').count().orderBy('Crime Date')

display(q3_result)
```

```
# SF downtown spacial range: X (-122.4213,-122.4313), Y(37.7540,37.7740). 
# Spark SQL based solution for Q3
CrimeDT = spark.sql("SELECT Date,DayOfWeek,COUNT(*) AS Count FROM sf_crime \
                      WHERE DayOfWeek = 'Sunday' AND \
                     X>=-122.4313 AND X <= -122.4213 AND Y<=37.7740 AND Y>=37.7540 \
                     GROUP BY Date,DayOfWeek \
                     ORDER BY Date ")
display(CrimeDT)
```

```
# Spark Dataframe based solution for Q4
# extract month and year from date
CrimeMoY= SF_DT.withColumn('month',month(SF_DT['Crime Date'])).withColumn('year', year(SF_DT['Crime Date']))
years = [2015, 2016, 2017, 2018]
CrimeMoY = CrimeMoY[CrimeMoY.year.isin(years)]
#display(CrimeMoY)

# count # crimes in each month for each year 
q4_result= CrimeMoY.groupby(['year','month']).count().orderBy('year','month')
display(q4_result)
```

```
# Spark SQL based solution for Q4
crimeMY= spark.sql("SELECT SUBSTRING(Date,7,4) AS Year,SUBSTRING(Date,1,2) AS Month, COUNT(*) AS Count \
                    FROM sf_crime \
                    GROUP BY Year, Month \
                    HAVING Year in (2015, 2016,2017, 2018) \
                    ORDER BY Year, Month")
display(crimeMY)
```

```
# Number of crimes in 2015,2016,and 2017 are constantly high, but in 2018, the nuber of crimes starts decreasing. This means that the SF PD have done some controls to help with the high crime rates and is effective. 
```

```
# park Dataframe based solution for Q5

from pyspark.sql.functions import to_timestamp
# Trans time to date formate
CrimeHours = SF_DT.withColumn('Time', to_timestamp(SF_DT['Time'],'HH:mm')) 
# get hour as a single column
CrimeHours = CrimeHours.withColumn('Hour',hour(CrimeHours['Time']))

display(CrimeHours)
```

```
# number of crime w.r.t the hour in certian day like 2015/12/15, 2016/12/15, 2017/12/15.
dates = ['12/15/2015','12/15/2016','12/15/2017']
crime_days = CrimeHours[CrimeHours.Date.isin(dates)]
q5_result = crime_days.groupby('Date','Hour').count().orderBy('Date','Hour')
display(q5_result)
```

```
# Spark SQL based solution for Q5
CrimeHour= spark.sql("SELECT SUBSTRING(Date,1,5) AS Day , SUBSTRING(Time,1,2) AS Hour,  COUNT(*) AS Count \
                       FROM sf_crime \
                       WHERE  SUBSTRING(Date,1,5) = '12/15' \
                       GROUP BY  Day, Hour \
                       ORDER BY Hour")
display(CrimeHour)
```

```
# By horizontally comparison on a same date, year 2017 has much more number of crime on 12/15 than in 2015. By looking at the number of crime happend in one month, 2015 Dec, 01:00-07:00AM have less crimes, while 17:00-20:00 have more crimes.
```
